```python
inverted_image = ImageOps.invert(Image.open('nofuture.jpg'))
pytesseract.image_to_string(inverted_image)
```

```python
width = inverted_image.size[0]
height = inverted_image.size[1]

top_left_x = width / 6
top_left_y = height * 5/6
bottom_right_x = width * 5/6
bottom_right_y = height-2
print(top_left_x, top_left_y, bottom_right_x, bottom_right_y)
print("dimensions", inverted_image.size[0], inverted_image.size[1])
inv_img_crop = inverted_image.crop((top_left_x, top_left_y, bottom_right_x, bottom_right_y))
inv_img_crop
```
